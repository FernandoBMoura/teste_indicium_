```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile


diretorio_extracao = 'banvic_data'


dataframes = {}


with zipfile.ZipFile('banvic_data.zip', 'r') as zip_ref:
    arquivos = zip_ref.namelist()
    zip_ref.extractall(diretorio_extracao)


for arquivo in arquivos:
    if arquivo.endswith('.csv'):
        nome_tabela = arquivo.split(sep='.')[0]
        caminho_csv = f'{diretorio_extracao}/{arquivo}'
        dataframes[nome_tabela] = pd.read_csv(caminho_csv)



for nome_tabela, df in dataframes.items():
    globals()[f'{nome_tabela}_df'] = df
    print(f'{nome_tabela}_df ok!')
    print()

```

```python
class Verificador:
    def __init__(self, arquivos):
        self.arquivos = arquivos

    def shape(self):
        for arquivo in self.arquivos:
            if arquivo.endswith('.csv'):
                nome_tabela = arquivo.split(sep='.')[0]
                df = globals()[f'{nome_tabela}_df']
                print(f'{nome_tabela}_df.shape:')
                print()
                print(df.shape)
                print()
                print()


    def null_sum(self):
        for arquivo in self.arquivos:
            if arquivo.endswith('.csv'):
                nome_tabela = arquivo.split(sep='.')[0]
                df = globals()[f'{nome_tabela}_df']
                print(f'{nome_tabela}_df.null_sum:')
                print()
                print(df.isnull().sum())
                print()
                print()

    def describe(self):
        for arquivo in self.arquivos:
            if arquivo.endswith('.csv'):
                nome_tabela = arquivo.split(sep='.')[0]
                df = globals()[f'{nome_tabela}_df']
                print(f'{nome_tabela}_df.describe:')
                print()
                print(df.describe())
                print()
                print()
```
